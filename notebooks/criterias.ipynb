# Sklepy

```python
CITY = "Wroc≈Çaw, Poland"
```

```python
tag = {"shop": "supermarket"}
```

```python
from src.data_processing import get_city_boundary, get_osm_data

city_gdf = get_city_boundary(CITY)
objects = get_osm_data(CITY, tag)
```

```python
import folium


def visualization(objects):
    # Przekszta≈Çcamy do uk≈Çadu metrycznego (np. EPSG:3857)
    gdf_proj = city_gdf.to_crs(epsg=3857)

    # Obliczamy centroid na uk≈Çadzie metrycznym, a potem wracamy do WGS 84
    centroid = gdf_proj.geometry.centroid.to_crs(epsg=4326).iloc[0]

    m = folium.Map(location=[centroid.y, centroid.x], zoom_start=10)

    # Dodanie punkt√≥w na mapƒô
    for coord in objects:
        folium.Marker(location=[coord[0], coord[1]]).add_to(m)

    # Wy≈õwietlenie mapy
    return m

visualization(objects)
```

```python
tag = {"shop": "mall"}
objects = get_osm_data(CITY, tag)
visualization(objects)
```
